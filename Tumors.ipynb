```
training_path = '/content/tumors/Training'
testing_path = '/content/tumors/Testing'

categories = np.array(['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor'])

```

```

for category in categories:
  path = os.path.join(training_path, category)
  for img in os.listdir(path):
      img_array = cv2.imread(os.path.join(path, img))
      plt.imshow(img_array)
      plt.title(category)
      plt.show()
      plt.axis('off')
      break
  break
 
```

```
train_data = []

def create_training_data():
  for category in categories:
    path = os.path.join(training_path, category)
    class_num = categories.index(category)
    for img in os.listdir(path):
      try:
        img_array = cv2.imread(os.path.join(path, img))
        new_array = cv2.resize(img_array, (150, 150))
        train_data.append([new_array, class_num])
      except Exception as e:
        pass

create_training_data()
```

```
train_data[:1]
```

```
image = []
labels = []

for features, label in train_data:
  image.append(features)
  labels.append(label)

image = np.array(image)
print(image.shape)
labels = np.array(labels)
print(labels.shape)
```

```
image = image/255.0
```

Now let,s split our training images so that we also have some validation images in our models. Just to make sure that we do not overfit while training!

```
split = int(len(image) * 0.8)
train_images = image[:split]
train_labels = labels[:split]
validation_images = image[split:]
validation_labels = labels[split:]
```

```
print('Length Training Images: ', len(train_images))
print('Length Training Labels: ', len(train_labels))
print('Length Validation Images: ', len(validation_images))
print('Length Validation Labels: ', len(validation_labels))
print(train_images.shape)
```

Let's do the same data manipulation with our testing data!

```
test_data = []

def create_testing_data():
  for category in categories:
    path = os.path.join(testing_path, category)
    class_num = categories.index(category)
    for img in os.listdir(path):
      try:
        img_array = cv2.imread(os.path.join(path, img))
        new_array = cv2.resize(img_array, (150, 150))
        test_data.append([new_array, class_num])
      except Exception as e:
        pass

create_testing_data()
```

```
image_test = []
labels_test = []

for features, label in test_data:
  image_test.append(features)
  labels_test.append(label)

```

```
image_test = np.array(image_test)
image_test = image_test/255.0
```

```
print(image_test.shape)
print(labels_test.shape)
```

```
from keras.utils.np_utils import to_categorical # convert to one-hot-encoding. We want to do that since we have a multiclass labels. This will convert the array into a binary matrix.
#labels = to_categorical(labels, num_classes = 4)
train_labels = to_categorical(train_labels, num_classes = 4)
validation_labels = to_categorical(validation_labels, num_classes = 4)
```

```
print(train_labels.shape)
```

Alright! Model time. We will be using a CNN with multiple layers.

```
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), input_shape = (150, 150, 3), padding = 'same', activation = 'relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu', padding = 'same'),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(64, (3, 3), activation = 'relu', padding = 'same'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),

    tf.keras.layers.Conv2D(128, (3, 3), activation = 'relu', padding = 'same'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),

    tf.keras.layers.Conv2D(128, (3, 3), activation = 'relu', padding = 'same'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1024, activation = 'relu'),
    tf.keras.layers.Dense(1024, activation = 'relu'),
    tf.keras.layers.Dense(512, activation = 'relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(4, activation = 'softmax')
])

optimizer = tf.keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999)
model.compile(optimizer = optimizer,
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])

model.summary()
```

```
history = model.fit(train_images, train_labels, epochs = 50, validation_data = (validation_images, validation_labels))
```

```
loss = history.history['loss']
val_loss = history.history['val_loss']
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

epochs = range(50)

plt.figure(figsize = (10, 6))
# First Graph: Accuracy & Val_Accuracy
plt.subplot(1, 2, 1)
plt.plot(epochs, accuracy, label = 'Accuracy')
plt.plot(epochs, val_accuracy, label = 'Val Accuracy')
plt.legend(loc = 'upper right')
plt.title('Accuracy & Val Accuracy')

# Second Graph: Loss and Val_Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, loss, label = 'Loss')
plt.plot(epochs, val_loss, label = 'Val Loss')
plt.legend(loc = 'upper right')
plt.title('Loss & Val Loss')

plt.show()
```

```
model.save('brain_tumor_model.h5')
```

Let's now use our model to predict the images we have in our test dataset!

```
image_batch = image_test
label_batch = np.array(labels_test)
label_batch = np.array(labels_test)

predicted_batch = model.predict(image_batch)
predicted_ids = np.argmax(predicted_batch, axis = 1)
predicted_class_names = categories[predicted_ids]

print('True labels:      ', label_batch)
print('Predicted labels: ', predicted_ids)
```

```
plt.figure(figsize = (25, 20))
for i in range(len(image_batch)):
  plt.subplot(20, 20, i+1)
  plt.subplots_adjust(hspace = 1.0)
  plt.imshow(image_batch[i])
  color = 'blue' if predicted_ids[i] == label_batch[i] else 'red'
  plt.title(predicted_class_names[i].title(), color = color)
  plt.axis('off')
  plt.suptitle('Model Predictions: Blue is correct, red is incorrect')
```
